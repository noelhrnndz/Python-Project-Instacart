# Step 2 In this Exercise, you learned how to find the aggregated mean of the “order_number” column grouped by “department_id” for a subset of your dataframe. Now, repeat this process for the entire dataframe.


```python
# Find aggregated mean of the "order_number" column grouped by "department_id"

ords_prods_merge.groupby('department_id').agg({'order_number': ['mean']})
```

# Step 4 Follow the instructions in the Exercise for creating a loyalty flag for existing customers using the transform() and loc() functions.

```python
# First split data in groups
# Create 'max_order' column to show flag results
# order_prods_merge is grouped by 'user_id'
# transform() function is applied on 'order_number' with np.max

ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
#ords_prods_merge.head(100)
```

### To display all rows use:

```python
# pd.options.display.max_rows = None
# Did not display all rows to save RAM
```

```python
# After max_order column is created, set flag with loc()

ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Check frequency 
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
# Cross check multiple columns with .head()

ords_prods_merge[['user_id', 'loyalty_flag', 'order_number']].head(60)
```

# Step 5 The marketing team at Instacart wants to know whether there’s a difference between the spending habits of the three types of customers you identified. Use the loyalty flag you created and check the basic statistics of the product prices for each loyalty category (Loyal Customer, Regular Customer, and New Customer). What you’re trying to determine is whether the prices of products purchased by loyal customers differ from those purchased by regular or new customers.


```python
# Check basic statistics of product prices for each loyalty category

ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

# Step 6 The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by looking at the prices of the items people are buying.

```python
# Group 'user_id' & prices
# Create new column 'mean_price_spent
# transform() prices into mean value

ords_prods_merge['mean_price_spent'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Qucik check of results

ords_prods_merge[['user_id', 'mean_price_spent']].head(60)
```

```python
# Create a spending flag for each user based on the average price across all their orders as "Low spender" or "High spender"

ords_prods_merge.loc[ords_prods_merge['mean_price_spent'] < 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['mean_price_spent'] >= 10, 'spending_flag'] = 'High spender'
```

```python
# Frequency check

ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

# Step 7 In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the same item again), the Instacart team wants you to determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column.

```python
# Group
# Create new column 'median_days'
# transform() median

ords_prods_merge['median_days'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Quick check of results

ords_prods_merge[['user_id', 'days_since_prior_order', 'median_days']].head(60)
```

```python
# Create flag for “Non-frequent customer”, “Regular customer”, & “Frequent customer.”

ords_prods_merge.loc[ords_prods_merge['median_days'] > 20, 'frequent_customer_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_days'] > 10) & (ords_prods_merge['median_days'] <= 20), 'frequent_customer_flag'] = 'Regular customer' 
```

```python
ords_prods_merge.loc[ords_prods_merge['median_days'] <= 10, 'frequent_customer_flag'] = 'Frequent customer'
```

```python
# Frequency check

ords_prods_merge['frequent_customer_flag'].value_counts(dropna = False)
```

```python
# Cross checking multiple columns with .head()

ords_prods_merge[['user_id', 'days_since_prior_order', 'frequent_customer_flag']].head(60)
```

# Step 4. Wrangle the data so that it follows consistent logic; for example, rename columns with illogical names and drop columns that don’t add anything to your analysis.


```python
# Rename First Name to first_name

df.rename(columns = {'First Name' : 'first_name'}, inplace = True)

# Rename Surenam to Last Name

df.rename(columns = {'Surnam' : 'last_name'}, inplace = True)

#Rename Gender to gender

df.rename(columns = {'Gender' : 'gender'}, inplace = True)

# Rename STATE to state

df.rename(columns = {'STATE' : 'state'}, inplace = True)

# Rename Age to age

df.rename(columns = {'Age' : 'age'}, inplace = True)

# Rename n_dependants to children

df.rename(columns = {'n_dependants' : 'children'}, inplace = True)

# Drop fam_status, not necessary
df.drop(columns = ['fam_status'])
```
